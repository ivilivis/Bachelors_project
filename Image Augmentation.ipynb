```python
#For converting color images only!!

import os
from PIL import Image, ImageFilter
import matplotlib as mpl
import matplotlib.cm as mtpltcm
import matplotlib.pyplot as plt

file_path = r"D:\Datasets FAces\test"

# traverse root directory, and list directories as dirs and files as files
for root, dirs, files in os.walk(file_path):
    path = root.split(os.sep)
    #print((len(path) - 1) * '---', os.path.basename(root))
    for file in files:
        new_path = file_path + "\\" + file
        image = Image.open(new_path)
        #print(new_path + " Loaded")
        
        # Blurring the image
        #imgBlur = image.filter(ImageFilter.BoxBlur(6))
        #print(new_path + " is blurred")
        
        # save a image using extension
        #imgBlur.save(new_path + "_blurred.jpg")
        
        
        #Create Greyscale
        imgGray = image.convert('L')
        imgGrayBlur = imgGray.filter(ImageFilter.BoxBlur(6))
        imgGrayBlur.save(new_path + "_grey_blurred.jpg")
        
        #Create Greyscale 2 for second run
        #imgGray2 = image.convert('L')
        #imgGray2.save(new_path + "_grey2.jpg")
        
        #initialize the colormap (jet)
        colormap = mpl.cm.jet
        #add a normalization
        cNorm = mpl.colors.Normalize(vmin=0, vmax=255)
        #init the mapping
        scalarMap = mtpltcm.ScalarMappable(norm=cNorm, cmap=colormap)
        colors = scalarMap.to_rgba(imgGrayBlur)
        #colors.save(new_path + "_fake_thermal.jpg")
        imgplot = plt.imshow(colors)
        plt.axis('off')
        imgplot.figure.savefig(new_path + "_fake_thermal.jpg", bbox_inches='tight', pad_inches=0)   # save the figure to file
        
        #print(new_path + " is saved")
```

```python
#For converting Thermal images only!!

import os
from PIL import Image, ImageFilter
import matplotlib as mpl
import matplotlib.cm as mtpltcm
import matplotlib.pyplot as plt

image_no = 0
file_path = r"D:\Datasets FAces\test"
image_set = file_path[-1]

# traverse root directory, and list directories as dirs and files as files
for root, dirs, files in os.walk(file_path):
    path = root.split(os.sep)
    #print((len(path) - 1) * '---', os.path.basename(root))
    for file in files:
        new_path = file_path + "\\" + file
        image = Image.open(new_path)
        #print(new_path + " Loaded")
        image.save(new_path + str(image_set) + str(image_no) + "_stock.jpg")
        
        print(new_path)
        
        # Blurring the image
        #imgBlur = image.filter(ImageFilter.BoxBlur(6))
        #print(new_path + " is blurred")
        
        # save a image using extension
        #imgBlur.save(new_path + "_blurred.jpg")
        
        
        #Create Greyscale
        imgGray = image.convert('L')
        imgGray.save(new_path + str(image_set) +str(image_no) + "_grey.jpg")
        
        #initialize the colormap (jet)
        colormap = mpl.cm.jet
        #add a normalization
        cNorm = mpl.colors.Normalize(vmin=0, vmax=255)
        #init the mapping
        scalarMap = mtpltcm.ScalarMappable(norm=cNorm, cmap=colormap)
        colors = scalarMap.to_rgba(imgGray)
        #colors.save(new_path + "_fake_thermal.jpg")
        imgplot = plt.imshow(colors)
        plt.axis('off')

        imgplot.figure.savefig(new_path + str(image_set) +str(image_no) + "_fake_rgb.jpg",bbox_inches='tight',pad_inches = 0)   # save the figure to file
        
        print(new_path + " is saved")
        image_no += 1
```
